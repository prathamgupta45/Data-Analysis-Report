## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
user_details = pd.read_excel('Data Analyst Intern Assignment - Excel.xlsx',"UserDetails.csv")
cooking_sessions = pd.read_excel('Data Analyst Intern Assignment - Excel.xlsx',"CookingSessions.csv")
order_details = pd.read_excel('Data Analyst Intern Assignment - Excel.xlsx',"OrderDetails.csv")
```

## Data cleaning and processing

```python
user_details.fillna({'Age': user_details['Age'].mean()}, inplace=True)
cooking_sessions.dropna(subset=['Session ID'], inplace=True)
```

```python
user_details.columns = user_details.columns.str.lower()
cooking_sessions.columns = cooking_sessions.columns.str.lower()
order_details.columns = order_details.columns.str.lower()
```

## Merge the files

```python
merged_data = pd.merge(user_details, cooking_sessions, on="user id", how="inner")
```

```python
final_data = pd.merge(merged_data, order_details, on="user id", how="inner")
```

```python
final_data.to_csv('Final Data.csv')
```

## Data Visualiztion

### Top Popular Dishes

```python
popular_dishes = final_data['dish name_y'].value_counts().head(10)
popular_dishes.plot(kind='bar')
plt.xlabel("Dish Names")
plt.title("Top 10 Popular Dishes")
plt.show()
```

### Average order by Age group

```python
age_group_orders = final_data.groupby('age')['total orders'].mean()
age_group_orders.plot(kind='bar')
plt.title("Average Orders by Age Group")
plt.show()
```

### Order Value Distribution

```python
final_data['amount (usd)'].plot(kind='hist', bins=20, color='lightgreen', edgecolor='black')
plt.title("Order Value Distribution")
plt.xlabel("Order Value")
plt.ylabel("Frequency")
plt.show()
```

### Number of participations by age group

```python
sns.countplot(x='age', data=final_data, palette='pastel')
plt.title("Cooking Session Participation by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Participants")
plt.xticks(rotation=45)
plt.show()
```

### Number of order by meal type

```python
meal_type_counts = final_data['meal type_y'].value_counts()
meal_type_counts.plot(kind='bar', color='teal', edgecolor='black')
plt.title("Number of Orders by Meal Type")
plt.xlabel("Meal Type")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.show()
```
